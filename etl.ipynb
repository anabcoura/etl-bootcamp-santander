```python
from google import genai
from google.genai import types

client = genai.Client(
    api_key= genai_api_key
)    


def generate_ai_news(name):
    response = client.models.generate_content(
    model="gemini-2.5-flash",
    config=types.GenerateContentConfig(
        system_instruction="Você é um especialista em marketing bancário."),
    contents= f"Crie uma mensagem para {name} sobre a importância dos investimentos (máximo de 100 caracteres)" )
    return response.text

news_list = []

for i, row in df.iterrows():
    user_id = row['UserID']
    user_name = row['Nome']
    news_content = generate_ai_news(user_name)
    entry = {
        "id": int(user_id),
        "nome": user_name,
        "mensagem": news_content
    }
    news_list.append(entry)
```

```python
import json

with open('ai_news.json', 'w', encoding='utf-8') as json_file:
    json.dump(news_list, json_file, indent=4, ensure_ascii=False)
```
