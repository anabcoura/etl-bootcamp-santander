```python
import openai
from openai import OpenAI

client = OpenAI(
    api_key = openai_api_key
)

def generate_ai_news(user_names):
    completion = client.chat.completions.create(model="gpt-3.5-turbo", messages=[
        {"role": "system", "content": "Você é um especialista em marketing bancário"},
        {"role": "user", "content": f"Crie uma mensagem para {user_names} sobre a importância dos investimentos (máximo de 100 caracteres)"}])

    return completion.choices[0].message.content

for user in users_names:
    news = generate_ai_news(user)
    print(news)
```
